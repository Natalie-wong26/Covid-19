## Contoh : Sheet "Kasus Aktif"

```python
data = 'data/COVID-19 Indonesia.xlsx'
daily_cases=pd.read_excel(data,sheet_name='Kasus Aktif')

daily_cases.head()
```

```python
#Handling Missing Value
daily_cases.fillna(0,inplace=True)
daily_cases.set_index('Date',inplace=True)
daily_cases.head()
```

```python
# plot grafik pertumbuhan jakarta

daily_cases['Jakarta'].plot();
plt.title("Grafik kasus per hari DKI Jakarta");
```

### Jawablah Pertanyaan Berikut dan Ceritakan dalam Bentuk Grafik / Visualisasi

- Bagaimana perkembangan covid-19 di Indonesia?
- Apakah kita sudah mengalami puncak kedua?
- Provinsi mana yang paling banyak total kasusnya? (bandingkan hingga 15 provinsi teratas)
- Apakah PPKM sudah cukup efektif mengurangi kasus Covid-19?
- Bagaimana perkembangan vaksinasi di Indonesia baik dosis pertama atau kedua?

```python
daily_cases.info()
```

```python
daily_cases['Indonesia'] = daily_cases.sum(axis = 1)
daily_cases.head(3)
```

```python
sns.set_style('whitegrid')
fig, axes = plt.subplots(figsize=(20,6))

ax = sns.lineplot(data=daily_cases, x='Date', y='Indonesia', color = 'red')

ax.xaxis.set_major_formatter(mdates.DateFormatter('%b-%y'))
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))

ax.set_xlabel("")
ax.set_ylabel("")

# Set tick font size
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(15)

ax.set_title("Perkembangan Kasus Aktif per Hari di Indonesia\n15 Maret 2020 s/d 23 Agustus 2021", fontsize = 20, pad = 18)


fig.tight_layout()
```

Pada grafik terlihat bahwa kasus covid-19 aktif per hari di Indonesia mengalami kenaikan seiring waktu dimana gelombang pertama terjadi mulai dari November 2020 sampai Januari 2021 dengan puncaknya berada di Januari 2020. Gelombang kedua dimulai dari Mei 2021 sampai Juli 2021 yang puncaknya adalah pada pertengahan Juli 2021. Pada gelombang kedua terdapat kenaikan yang sangat drastis dibandingkan dengan gelombang pertama.

```python
gel_satu = daily_cases.loc[(daily_cases.index >= '2020-11-01') & (daily_cases.index <= '2021-01-31')]
#gel_satu.head(3)
```

```python
gel_dua = daily_cases.loc[(daily_cases.index >= '2021-05-01') & (daily_cases.index <= '2021-07-31')]
#gel_dua.tail(3)
```

```python
puncak_gel_satu = "{:,.0f}".format(gel_satu['Indonesia'].max())
puncak_gel_dua = "{:,.0f}".format(gel_dua['Indonesia'].max())

#dtype tanggalnya adalah Timestamp
tgl_puncak_satu = gel_satu['Indonesia'].idxmax().strftime('%d-%m-%Y')
tgl_puncak_dua = gel_dua['Indonesia'].idxmax().strftime('%d-%m-%Y')

print(f"Puncak kasus aktif gelombang pertama adalah pada {tgl_puncak_satu}, yaitu sebanyak {puncak_gel_satu} kasus.")
print(f"Puncak kasus aktif gelombang kedua adalah pada {tgl_puncak_dua}, yaitu sebanyak {puncak_gel_dua} kasus.")

persen = "{:.2%}".format(gel_dua['Indonesia'].max()/gel_satu['Indonesia'].max())
print(f"Terdapat kenaikan sebanyak {persen} pada puncak gelombang ke dua terhadap puncak kasus gelombang satu.")

print("Bertepatan dengan Pemberlakuan Pembatasan Kegiatan Masyarakat (PPKM) Darurat yang dimulai 3 Juli 2021 sampai 25 Juli 2021, laju kasus covid-19 terlihat menurun mulai pada pertengahan bulan Juli. Akan tetapi, diperlukan analisis lebih lanjut untuk menentukan apakah PPKM tersebut memang efektif.")
```

```python
#mengubah wide data menjadi long data
daily_cases_melted = daily_cases.copy()
daily_cases_melted = daily_cases.melt(var_name='Provinsi', value_name='Kasus', ignore_index = False)
daily_cases_melted.sample(5)
```

```python
daily_cases_melted = daily_cases_melted.groupby('Provinsi', as_index=False).sum().sort_values('Kasus', ascending=False, ignore_index=True)
daily_cases_melted.head(3)
```

```python
#provinsi menjadi index, dataframe daily_cases_melted hanya memiliki 1 kolom , drop row pertama karena adalah data seluruh Indonesia
daily_cases_melted.info()
```

```python
#overwrite dataframe daily_cases_melted tanpa baris 'Indonesia'
daily_cases_melted = daily_cases_melted.iloc[1:]
```

```python
sns.set_style('whitegrid')
fig, axes = plt.subplots(figsize=(14,6))

ax = sns.barplot(data=daily_cases_melted.head(15), x='Kasus', y='Provinsi', palette = ("rocket"))

ax.set_xlabel("")
ax.set_ylabel("")

# Set tick font size
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(15)
    
ax.set_title("Kasus Aktif per Provinsi", fontsize = 20, pad = 18)


fig.tight_layout()
```

## Import Sheet `Statistik_Harian`

```python
data = 'data/COVID-19 Indonesia.xlsx'
statistik_harian=pd.read_excel(data,sheet_name='Statistik Harian')

statistik_harian.tail(3)
```

```python
statistik_harian.dropna(subset=['Date'], inplace=True)
#statistik_harian.info()
```

```python
statistik_harian_clean = statistik_harian.iloc[:,[0,42,43,44,53,54,55,58,59,62,63,66,67,68,69,70,71]]
```

```python
statistik_harian_clean.fillna(0,inplace=True)
statistik_harian_clean.set_index('Date',inplace=True)
# drop row jika semua columns = 0
statistik_harian_clean = statistik_harian_clean.loc[(statistik_harian_clean!=0).any(axis=1)]
```

```python
# Column Data yang tersedia setiap barisnya adalah sum atas semua baris diatasnya (perubahan nilai)
# untuk mencari dosis harian dan drop column awal (yang terdapat data perubahan nilai)
def harian(col):
    new_col = col + " harian"
    statistik_harian_clean[new_col] = statistik_harian_clean[col].diff().fillna(statistik_harian_clean[col])
    statistik_harian_clean.drop(columns=[col], inplace=True)

for column_name, _ in statistik_harian_clean.iteritems():
    harian(column_name)
```

```python
statistik_harian_clean.head()
```

```python
statistik_harian_clean.info()
```

```python
sns.set_style('whitegrid')
fig, axes = plt.subplots(figsize=(20,6))

ax = sns.lineplot(data=statistik_harian_clean, x='Date', y='Dosis pertama harian', color = 'red', label ='Dosis Pertama', linewidth = 3)
ax = sns.lineplot(data=statistik_harian_clean, x='Date', y='Dosis kedua harian', color = 'green', label ='Dosis Kedua', linewidth = 3)
ax = sns.lineplot(data=statistik_harian_clean, x='Date', y='Dosis ketiga harian', color = 'blue', label ='Doses Ketiga', linewidth = 3)

ax.xaxis.set_major_formatter(mdates.DateFormatter('%b-%y')) #major tick hanya bulan dan tahun
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1)) #interval per bulan

ax.set_xlabel("")
ax.set_ylabel("")

# ukuran font untuk tick
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(15)

# agar angka tidak muncul dalam scientific format
plt.ticklabel_format(style='plain', axis='y')
    
ax.set_title("Perkembangan Vaksinasi di Indonesia", fontsize = 20, pad = 18)

fig.tight_layout()
plt.setp(ax.get_legend().get_texts(), fontsize='22') # for legend text
```
